```python
import numpy as np
import matplotlib.pyplot as plt

# Keep this order (train order) for BOTH series
classes = [
    "Man","Clothing","Person","Tree","Human face","Footwear","Woman","Window","Flower","Wheel",
    "Human hair","Plant","Human arm","Car","Human head","Girl","Human body","Mammal","Tire","Chair",
]

train_counts = {
    "Man":4893, "Clothing":4781, "Person":4025, "Tree":3998, "Human face":3561, "Footwear":2768,
    "Woman":2610, "Window":2045, "Flower":1394, "Wheel":1376, "Human hair":963, "Plant":932,
    "Human arm":918, "Car":913, "Human head":830, "Girl":806, "Human body":753, "Mammal":721,
    "Tire":618, "Chair":611,
}

val_counts = {
    "Man":122, "Clothing":252, "Person":232, "Tree":95, "Human face":57, "Footwear":125,
    "Woman":52, "Window":213, "Flower":35, "Wheel":214, "Human hair":122, "Plant":82,
    "Human arm":142, "Car":127, "Human head":145, "Girl":43, "Human body":122, "Mammal":206,
    "Tire":120, "Chair":3,
}

train = np.array([train_counts.get(c, 0) for c in classes], dtype=float)
val   = np.array([val_counts.get(c, 0) for c in classes], dtype=float)

# "Normalized histogram": convert counts -> proportions that sum to 1
train_norm = train / train.sum()
val_norm   = val / val.sum()

x = np.arange(len(classes))
w = 0.42

fig, ax = plt.subplots(figsize=(14, 5))
ax.bar(x - w/2, train_norm, width=w, label="train", alpha=0.9)
ax.bar(x + w/2, val_norm,   width=w, label="val",   alpha=0.9)

ax.set_xticks(x)
ax.set_xticklabels(classes, rotation=45, ha="right")
ax.set_ylabel("Proportion of boxes (counts / total)")
ax.set_title("Train vs Val class distributions of 20 most common train classes")
ax.grid(axis="y", alpha=0.25)
ax.legend()
plt.tight_layout()
plt.show()
```
